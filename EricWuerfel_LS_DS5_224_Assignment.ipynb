```python
# 'country' 'name' 'world_6region'

# entities = entities.drop('alt_5', 'alternative_1', 'alternative_2', 'alternative_3',
#        'alternative_4_cdiac', 'arb1', 'arb2', 'arb3', 'arb4', 'arb5', 'arb6',
#        'g77_and_oecd_countries', 'gapminder_list', 'god_id', 'gwid',
#        'income_groups', 'is--country', 'iso3166_1_alpha2', 'iso3166_1_alpha3',
#        'iso3166_1_numeric', 'iso3166_2', 'landlocked', 'latitude', 'longitude',
#        'main_religion_2008', 'pandg', 'un_state',
#        'unicode_region_subtag', 'upper_case_name', 'world_4region',
#        )

entities_name_region = entities[['country', 'name', 'world_6region']]

entities_name_region.head()
```

## Merge Data

```python
df = income.merge(lifespan).merge(population)
```

```python
df.shape
```

```python
df.head()
```

```python
# Create a new DataFrame for Entity Information
entity_info = entities[['country', 'name', 'world_6region']]
```

```python
df.head()
```

```python
entity_info.head()
```

```python
# Merge Data DataFrame with Entity Info DataFrame

# Stitch them together: left df --> geo <--> country <-- entity_info
df = pd.merge(df, entity_info, left_on='geo', right_on='country')
print(df.shape)
df.head()
```

```python
# Now we can get rid of the tags we used to merge the DataFrames
df = df.drop(columns=['geo', 'country'], axis='columns')
df.head()
```

```python
df = df.rename(columns={
    'time':'year'
    , 'income_per_person_gdppercapita_ppp_inflation_adjusted':'income'
    , 'life_expectancy_years':'lifespan'
    , 'population_total':'Population'
    , 'world_6region':'Region'
    , 'name':'country'
})
```

```python
df['region'] = df['region'].replace('_', ' ', regex=True).str.title()
```

```python
df.head()
```

## Plot Visualization

```python
present = df[df['year'] == 2018]
present.head()
```

```python
# Seaborn RelPlot
sns.relplot(x='income', y='lifespan', hue='region', size='population',
          data=present, legend='brief');

# Set Title
plt.title('An informative title')

# Use Log Scale
plt.xscale('log');

# Set ranges for the axes
plt.xlim(1, 150000);
plt.ylim(0, 100);

# Set Labels
plt.xlabel('Income')
plt.ylabel('Life Expectancy')
```
