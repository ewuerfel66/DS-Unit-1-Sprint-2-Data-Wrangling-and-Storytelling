## What are the properties of your data?
- Is your primary variable of interest continuous or discrete?
- Is in wide or long (tidy) format?
- Does your visualization involve multiple variables?
- How many dimensions do you need to include on your plot?

Can you express the main idea of your visualization in a single sentence?

How hard does your visualization make the user work in order to draw the intended conclusion?

# Cable News Mentions

From: [Which 2020 Candidate Was Mentioned Most on Cable News Last Week](https://fivethirtyeight.com/features/which-2020-candidate-was-mentioned-most-on-cable-news-last-week-2/)

Data Available [Here](https://github.com/fivethirtyeight/data/tree/master/media-mentions-2020)

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/media-mentions-2020/cable_weekly.csv')
```

```python
df['date'] = df['date'].apply(str)
```

```python
py_dates = []

for index, row in df.iterrows():
  py_dates.append(parser.parse(df['date'][index]))
```

```python
df['py_dates'] = py_dates
```

## Arrange the data into a new dataframe

```python
plot_data = pd.DataFrame()
```

```python
unique_names = []

for index, row in df.iterrows():
  if df['name'][index] not in unique_names:
    unique_names.append(df['name'][index])
    
plot_data['names'] = unique_names
```

```python
week1_Mentions = []
week2_Mentions = []
```

```python
def getMentions_week1(candidate):
  for index, row in df.iterrows():
    if df['name'][index] == candidate:
      if df['py_dates'][index] == datetime(2019, 5, 19, 0, 0):
        week1_Mentions.append(df['matched_clips'][index])
```

```python
def getMentions_week2(candidate):
  for index, row in df.iterrows():
    if df['name'][index] == candidate:
      if df['py_dates'][index] == datetime(2019, 5, 26, 0, 0):
        week2_Mentions.append(df['matched_clips'][index])
```

```python
for _ in unique_names:
  getMentions_week1(_)
  getMentions_week2(_)
```

```python
plot_data['week1'] = week1_Mentions
plot_data['week2'] = week2_Mentions
```

```python
plot_data.head()
```

```python
plot_data = plot_data.sort_values(by='week2', ascending=False)
```

## Plot

```python
sns.set(style="whitegrid")

fig1, ax1 = plt.subplots()

sns.barplot(x="week1", y="names", data=plot_data,
           label="Week 1", color='#666666');

ax1.set(xlim=(0, 1800), ylabel="",
       xlabel="Mentions on Cable News")
sns.despine(left=True, bottom=True)
```

```python
sns.set(style="whitegrid")

fig2, ax2 = plt.subplots()

sns.barplot(x="week2", y="names", data=plot_data,
           label="Week 2", color='#ED713A');

ax2.set(xlim=(0, 1800), ylabel="",
       xlabel="Mentions on Cable News")
sns.despine(left=True, bottom=True)
```
